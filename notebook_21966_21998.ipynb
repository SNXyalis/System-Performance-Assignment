```python
import threading
bacc_A, bacc_B = 100_000, 100_000

def transfer5(lock):
  global bacc_A, bacc_B
  for _ in range(100000): 
    lock.acquire()
    bacc_A-=5
    bacc_B+=5
    lock.release()

def transfer10(lock):
  global bacc_A, bacc_B
  for _ in range(50000):
    lock.acquire()
    bacc_B-=10
    bacc_A+=10
    lock.release()
    
def transfer20(lock):
  global bacc_A, bacc_B
  for _ in range(100000):
    lock.acquire()
    bacc_A-=20
    bacc_B+=20
    lock.release()

def transfer40(lock):
  global bacc_A, bacc_B
  for _ in range(50000):
    lock.acquire()
    bacc_B-=40
    bacc_A+=40
    lock.release()


def main():
  

   lock = threading.Lock()
   t1 = threading.Thread(target = transfer5, args = (lock,))
   t2 = threading.Thread(target = transfer10, args = (lock,))
   t3 = threading.Thread(target = transfer20, args = (lock,))
   t4 = threading.Thread(target = transfer40, args = (lock,))

   t1.start()
   t2.start()
   t3.start()
   t4.start()


   t1.join()
   t2.join()
   t3.join()
   t4.join()


if __name__ == "__main__":
  main()
  print("A: %d B: %d " %(bacc_A,bacc_B))

```

```python
class Accounts:
  def __init__(self, amount,name):
    self.amount = amount
    self.name = name

  def deposit(self, amount):
    self.amount += amount

  def withdraw(self, amount):
    self.amount -= amount

  def transfer(from_account, to_account, amount,times,lock):
    for i in range(times):
       lock.acquire()
       from_account.withdraw(amount)
       to_account.deposit(amount)
       lock.release()

  def get_balance(self):
    return self.amount





def main():

   num1=Accounts(100_000,'bacc_A')
   num2=Accounts(100_000,'bacc_B')

   lock = threading.Lock()
   t1 = threading.Thread(target = Accounts.transfer, args = (num1,num2,5,100000,lock))
   t2 = threading.Thread(target = Accounts.transfer, args = (num2,num1,10,50000,lock))
   t3 = threading.Thread(target = Accounts.transfer, args = (num1,num2,20,100000,lock))
   t4 = threading.Thread(target = Accounts.transfer, args = (num2,num1,40,50000,lock))

   t1.start()
   t2.start()
   t3.start()
   t4.start()


   t1.join()
   t2.join()
   t3.join()
   t4.join()

   print(num1.get_balance())
   print(num2.get_balance())

  

   
if __name__ == "__main__":
  main()



```
